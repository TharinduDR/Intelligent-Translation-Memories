```python
import pandas as pd
```

```python
infersent_result = pd.read_csv("result/ES-ES/infersent.tsv", sep="\t")
infersent_result[:10]
```

```python
sbert_result = pd.read_csv("result/ES-ES/sbert.tsv", sep="\t")
sbert_result[:10]
```

```python
universal_encoder_result = pd.read_csv("result/ES-ES/universal_encoder.tsv", sep="\t")
universal_encoder_result[:10]
```

```python
infersent_result_sorted = infersent_result.sort_values('meteor_scores')
infersent_result_sorted = infersent_result_sorted.reset_index(drop=True)
infersent_result_sorted['infersent_meteor_scores'] = infersent_result_sorted['meteor_scores']
infersent_result_sorted[:10]
```

```python
sbert_result_sorted = sbert_result.sort_values('meteor_scores')
sbert_result_sorted = sbert_result_sorted.reset_index(drop=True)
sbert_result_sorted['sbert_meteor_scores'] = sbert_result_sorted['meteor_scores']
sbert_result_sorted[:10]
```

```python
universal_encoder_result_sorted = universal_encoder_result.sort_values('meteor_scores')
universal_encoder_result_sorted = universal_encoder_result_sorted.reset_index(drop=True)
universal_encoder_result_sorted['universal_encoder_meteor_scores'] = universal_encoder_result_sorted['meteor_scores']
universal_encoder_result_sorted[:10]
```

```python
ax = infersent_result_sorted.plot.line(y=['infersent_meteor_scores'])
sbert_result_sorted.plot.line(ax=ax, y=['sbert_meteor_scores'])
universal_encoder_result_sorted.plot.line(ax=ax, y=['universal_encoder_meteor_scores'])
```

```python
analysis_full_df = infersent_result[['source','target']]
analysis_full_df["infersent_meteor_scores"] = infersent_result['meteor_scores']
analysis_full_df["sbert_meteor_scores"] = sbert_result['meteor_scores']
analysis_full_df["universal_encoder_meteor_scores"] = universal_encoder_result['meteor_scores']
analysis_full_df[:10]
```

```python
different_rows = analysis_full_df.loc[analysis_full_df['infersent_meteor_scores'] != analysis_full_df['sbert_meteor_scores']]
different_rows[:10]
```
